## ④	결측치가 없는 컬럼는 제외하고 시각화(정렬전 bar plot, barh plot, 정렬 후 bar plot, barh plot).

```python
# missing_cnt : 열별 결측치 갯수
only_missing_cnt = missing_cnt[missing_cnt!=0]
only_missing_cnt
```

```python
# only_missing_cnt(결측치가 있는 열 갯수)를 barplot
only_missing_cnt.plot(kind = 'bar', 
                      rot = 30, 
                      figsize = (10,5),
                      ylim = [0,650000]
                     )
for i, v in enumerate(only_missing_cnt):
    if v<10 :
        plt.text(i-.05, v+5000, v)
    elif  v < 100:
        plt.text(i-.1, v+5000, v)
    elif v < 1000:
        plt.text(i-.25, v+5000, v)
    else:
        plt.text(i-.4, v+5000, v)
plt.show()
```

```python
# only_missing_cnt(결측치가 있는 열 갯수)를 barhplot
only_missing_cnt.plot(kind = 'barh',
                      figsize = (10,5),
                      xlim = [0,650000]
                     )
for i, v in enumerate(only_missing_cnt):
    plt.text(v+5000, i-.2, v)
plt.show()
```

```python
# only_missing_cnt(결측치가 있는 열 갯수)를 barplot
plt.figure(figsize=(20,5))
g = sns.barplot(
    data=only_missing_cnt.to_frame(),
    x = only_missing_cnt.index,
    y = only_missing_cnt)
for i, v in enumerate(only_missing_cnt):
    if v<10 :
        plt.text(i-.05, v+5000, v, color='b')
    elif  v < 100:
        plt.text(i-.075, v+5000, v, color='b')
    elif v < 1000:
        plt.text(i-.1, v+5000, v, color='b')
    else:
        plt.text(i-.2, v+5000, v, color='b')
# plt.title('결측치 갯수')
# g.set(title='결측치 갯수')
g.set_title('결측치 갯수', color='r')
plt.show()
```

```python
only_missing_cnt_sorted = only_missing_cnt.sort_values(ascending=False)
only_missing_cnt_sorted
```

```python
only_missing_cnt_sorted.plot(kind='bar',
                             rot=20,
                             figsize = (15,5),
                             ylim=[0,650000],
                             yticks=[0,100000,200000,400000,500000,550000,600000,650000]
                            )
for i, v in enumerate(only_missing_cnt_sorted):
    if v<10 :
        plt.text(i-.05, v+5000, v)
    elif  v < 100:
        plt.text(i-.075, v+5000, v)
    elif v < 1000:
        plt.text(i-.15, v+5000, v)
    elif v < 100000:
        plt.text(i-.225, v+5000, v)
    else:
        plt.text(i-.35, v+5000, v, color='r', fontweight='bold', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(15,5))
g = sns.barplot(data=only_missing_cnt_sorted.to_frame(),
                 x = only_missing_cnt_sorted,
                 y = only_missing_cnt_sorted.index,
                )
for i, v in enumerate(only_missing_cnt_sorted):
    if v < 1000:
        plt.text(v+5000, i+.15, v)
    else:
        plt.text(v+5000, i+.2, v, color='r', fontweight='bold', fontsize=12)
g.set(xlim=[0,700000],
      xticks=[0,100000,200000,400000,500000,550000,600000,650000]
     )
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

# 4.	df 데이터 셋에서 사용하지 않을 컬럼은 제거
## ①	결측치가 너무 많은 컬럼은 제거한다. 결측치가 너무 많은 상위 9개 컬럼은 제거함.

```python
not_use = df.isnull().sum().sort_values(ascending=False).head(9)
not_use.index
```

```python
not_use['호정보'] / df.shape[0] * 100
```

```python
# not_use 컬럼들의 결측치 비율
round(not_use / df.shape[0] * 100, 2)
```

```python
# not_use 컬럼들의 결측치 비율(표준산업분류명  0.02%) - 리스트컴프리헨션
for col, percent in zip(not_use.index,[str(round(x/df.shape[0]*100, 2))+'%' for x in not_use]):
    print(col,percent)
```

```python
# not_use 컬럼들의 결측치 비율(표준산업분류명  0.02%) - map
not_use.map(lambda x : str(round(x/df.shape[0]*100, 2))+'%')
```

```python
# not_use 컬럼들의 결측치 비율(표준산업분류명  0.02%) - apply
not_use.apply(lambda x : str(round(x/df.shape[0]*100, 2))+'%')
```

```python
not_use.index # 삭제할 컬럼명
```

```python
df.info()
```

```python
#컬럼 삭제전 메모리 사용량(181.5+ MB), 컬럼수
print('컬럼 제거 전 shape :', df.shape)
df.drop(not_use.index, axis=1, inplace=True)
#컬럼 삭제후 메모리 사용량(139.6+ MB), 컬럼수
print('컬럼 제거 후 shape :', df.shape)
```

## ②	컬럼명에 “코드”나 “번호”가 있는 컬럼은 분석에 사용되지 않을 예정입니다. 제거함

```python
# 컬럼명
cols = df.columns
cols
```

```python
print('컬럼 제거 전 shape :', df.shape)
df.drop(drop_cols, axis=1, inplace=True)
print('컬럼 제거 후 shape :', df.shape)
```
