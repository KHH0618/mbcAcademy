# 2. 사이킷런(sklearn)이용
- 원핫인코딩을 하지 않고 라벨인코딩까지만 해야 작동. 원핫인코딩은 작동하지 않음

```python
from sklearn import datasets # iris 데이터
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
# 1. 데이터
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
print(iris_X[::50], iris_y[::50])
train_X, test_X, train_y, test_y = train_test_split(iris_X, iris_y,
                                                    test_size=0.1, # 기본값은 train:test = 0.75:0.25
                                                    random_state=4,
                                                    stratify=iris_y)
print(train_X.shape, test_X.shape, train_y.shape, test_y.shape)
```

```python
# 사이킷런에서 제공하는 다층퍼셉트론 기반의 분류 알고리즘 api
mlp = MLPClassifier(hidden_layer_sizes=(80,50,30),
                    activation='relu',
                    solver='adam', # adam이 기본값, optimizer
                    max_iter=1000, # 최대학습횟수
                    early_stopping=True, # 조기종료 활성화
                    n_iter_no_change=10, # 검증셋 손실함수가 개선되지 않으면 조기종료, 기본값 10
                    validation_fraction=0.1) # 검증셋 비율, 기본값 0.1
```

```python
mlp.fit(train_X, train_y)
```

```python
mlp.score(test_X, test_y)
```

```python
test_X[0], test_y[0]
```

```python
mlp.predict(test_X[0:1])
```

# 3. 클래스를 생성

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

```python
class DNNClassifier:
    @staticmethod
    def build(input_dim=4, activation='relu', output_dim=3):
#         model = Sequential()
#         model.add(Input(input_dim))
#         model.add(Dense(50, activation=activation))
#         model.add(Dense(30, activation=activation))
#         model.add(Dense(output_dim, activation='softmax'))
        model = Sequential([
            Input(input_dim),
            Dense(50, activation=activation),
            Dense(30, activation=activation),
            Dense(output_dim, activation='softmax')
        ])
        # 학습 과정 설정
        model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        return model
```

```python
# 1. 데이터 생성 및 전처리
# iris_Y = pd.get_dummies(iris_y).values
# train_X, test_X, train_Y, test_Y = train_test_split(iris_X, iris_Y, # 원핫인코딩한 데이터로 분류
#                                                     test_size=0.1, # 기본값은 train:test = 0.75:0.25
#                                                     random_state=4,
#                                                     stratify=iris_y)
# print(train_X.shape, test_X.shape, train_Y.shape, test_Y.shape)
train_Y = pd.get_dummies(train_y).values
test_Y = pd.get_dummies(test_y).values
# 2. 모델
model = DNNClassifier.build(input_dim=4, activation='relu', output_dim=3)
# 3. 모델학습
earlystopping = EarlyStopping(patience = 150)
hist = model.fit(train_X, train_Y, epochs=1000, validation_split=0.1, callbacks=[earlystopping])
```

```python
loss, accuracy = model.evaluate(test_X, test_Y)
print('정확도 : {}% loss : {}'.format(accuracy*100, loss))
```

# 4. 병렬처리

```python
from tensorflow.keras.layers import concatenate
from tensorflow.keras import Model
```

```python
test_X.shape, test_Y.shape, train_X.shape, train_Y.shape
```

```python
# 모델 구성
input_ = Input(shape=(4,))
dense1 = Dense(50, activation='relu')(input_)
dense2 = Dense(80, activation='relu')(input_)
dense3 = Dense(30, activation='relu')(input_)
x = concatenate([dense1, dense2, dense3])
# x = Dense(units=160, activation='relu')(input_)의 병렬처리 구현
output = Dense(units=3, activation='softmax')(x)
model = Model(inputs=input_, outputs=output)
model.summary()
```

```python
# 학습 정의(정의), 학습
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_X, train_Y, epochs=1000, verbose=1)
```

```python
# 학습평가
loss, accuracy = model.evaluate(test_X, test_Y)
print(accuracy*100,'%')
```
