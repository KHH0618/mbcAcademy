전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv (지역명, 평당분양가격, 연도, 월)
- 지역명2 : 지역명필드는 라벨인코딩하여 추가
- 평당분양가격n : 평당분양가격의 정규화 스케일 조정하여 추가
- 평당분양가격s : 평당분양가격의 표준화 스케일 조정하여 추가
- 연도n : 연도의 정규화 스케일 조정하여 추가
- 연도s : 연도의 표준화 스케일 조정하여 추가
- 월n : 월의 정규화 스케일 조정하여 추가
- 월s : 월의 표준화 스케일 조정하여 추가

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv(r'D:\KHH\ai\Download\shareData\전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv', encoding='cp949')
df = df.melt(id_vars=['지역'], var_name='연도월', value_name='평당분양가격')
df['연도'] = df['연도월'].apply(lambda x : int(x.split('년')[0]))
df['월'] = df['연도월'].apply(lambda x : int(x.split('년')[-1][:-1]))
df.sample(10)
```

```python
# 지역명2
le = LabelEncoder()
df['지역2'] = le.fit_transform(df['지역'])
df.head()
```

```python
# 평당분양가격n : 평당분양가격의 정규화 스케일 조정하여 추가

price_array = df['평당분양가격'].copy()
price_array = np.array(price_array)
price = MinMaxScaler()
price.fit(price_array.reshape(-1,1))
df['평당분양가격n'] = price.transform(price_array.reshape(-1,1))
df.head()
```

```python
# 평당분양가격s : 평당분양가격의 표준화 스케일 조정하여 추가
price_array_std = df['평당분양가격'].copy()
price_array_std = np.array(price_array_std)
price_std_scale = StandardScaler()
price_std_scale.fit(price_array_std.reshape(-1,1))
df['평당분양가격s'] = price_std_scale.transform(price_array_std.reshape(-1,1))
df.head()
```

```python
# 연도n : 연도의 정규화 스케일 조정하여 추가
year_array = df['연도'].copy()
year_array = np.array(year_array)
year_scale = MinMaxScaler()
year_scale.fit(year_array.reshape(-1,1))
df['연도n'] = year_scale.transform(year_array.reshape(-1,1))
df['연도n']
```

```python
# 연도s : 연도의 표준화 스케일 조정하여 추가
year_std_scale = StandardScaler()
year_std_scale.fit(year_array.reshape(-1,1))
df['연도s'] = year_std_scale.transform(year_array.reshape(-1,1))
df['연도s']
```

```python
# 월n : 월의 정규화 스케일 조정하여 추가
month_array = np.array(df['월']).reshape(-1,1)
month_scale = MinMaxScaler()
month_scale.fit(month_array)
df['월n'] = month_scale.transform(month_array)
df['월n']
```

```python
# 월s : 월의 표준화 스케일 조정하여 추가
month_std_scale = StandardScaler()
month_std_scale.fit(month_array)
df['월s'] = month_std_scale.transform(month_array)
df['월s']
```
